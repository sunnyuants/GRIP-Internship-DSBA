**What is Exploratory Data Analysis?**

Exploratory Data Analysis (EDA) is an approach that is used to analyze the data and discover trends, patterns, or check assumptions in data with the help of statistical summaries and graphical representations. 

**Here are six key steps that you can follow to conduct EDA:**

1.   Data Cleaning
2.   Data Collection
3.   Data Visualization
4.   Descriptive Statistics
5.   Data exploration
6.   Feature Engineering
7.   Hypothesis Testing
8.   Summarizing Findings

```python
ps=df.groupby("Ship Mode")[["Profit","Sales"]].sum()
ps
```

```python
import matplotlib.pyplot as plt
ps.plot(kind='bar',figsize=(6,4))
plt.title("Profit vs Sales",fontsize=19)
plt.ylabel("Profit vs Sales")
plt.show()
```

```python
cat=df["Category"].value_counts()
cat
```

```python
explode_list=[0,0.1,0.1]
colors_list=["Grey","red","blue"]
cat.plot(kind="pie",figsize=(15,5),autopct="%1.1f%%",shadow=True,startangle=180,explode=explode_list,labels=None,pctdistance=1.12,colors=colors_list)
plt.title("Category",fontsize="19")
plt.axis("equal")
plt.legend(labels=cat.index,loc="upper left")
plt.show()
```

```python
cs=df.groupby("Category").Sales.sum()
cp=df.groupby("Category").Profit.sum()
```

```python
ax0=cs.plot(kind="bar",figsize=(10,6),color="Pink")
ax0=cp.plot(kind="bar",figsize=(10,6),color="yellow")
ax0.set_title("Category wise Sales and Profit",fontsize=19)
ax0.set_ylabel("Total slaes and profit",fontsize=19)
ax0.set_xlabel("Category",fontsize=19)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x="Region",hue="Ship Mode",data=df)
plt.title("Region wise Ship Mode count")
plt.ylabel("Ship Mode count")
plt.show()
```

```python
df.groupby("Region")[["Sales","Profit"]].sum().sort_values(by=["Sales","Profit"],ascending=False).plot(kind="bar",color=["cornflowerblue","lightsalmon"])
plt.ylabel("Sales and Profit")
plt.xlabel("Region")
plt.title("Sales and Profit distribution among segments")
plt.show()
```

```python
reg=df["Region"].value_counts()
reg
```

```python
reg.plot(kind="pie",figsize=(7,7),autopct="%1.1f%%",
         shadow=True,labels=None,colors=["lightcoral","pink","lightblue","lightsalmon"])
plt.title("Region")
plt.legend(labels=reg.index,loc="upper left")
plt.axis("equal")
plt.show()
```
